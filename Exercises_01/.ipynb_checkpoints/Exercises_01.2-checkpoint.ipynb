# Introduzione

In questo esercizio verifichiamo l'esistenza di distribuzioni stabili, ossia distribuzioni per cui la combinazione lineare di due random variabili indipendenti ha la stessa distribuzione di quella di partenza a meno di un riscalamento. In particolare, verifichiamo che:
* Se consideriamo una distribuzione a varianza finita, come la distribuzione uniforme e quella esponenziale, allora il teorema del limite centrale ci assicura che la somma di estrazioni indipendenti converga ad una gaussiana.
* Se consideriamo invece una distribuzione lorentziana, la deviazione standard della distribuzione è infinita a cuasa delle code a leggi di potenza. Quindi il teorema del limite centrale non è più valido. Tuttavia Levy ha mostrato che esiste comunque una distribuzione stabile per la somma di estrazioni indipendenti che nel caso della lorentziana è la distribuzione di Cauchy.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import string
from scipy.optimize import curve_fit
from scipy.stats import norm
import matplotlib.mlab as mlab
from scipy.stats import cauchy

import seaborn as sns
sns.set_theme(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1.2, color_codes=True, rc=None)
```

```python
# Carico i dati dai file

file_uniform= "Exercises_01.2/file_uniform.txt"
file_exp= "Exercises_01.2/file_exp.txt"
file_lorentz = "Exercises_01.2/file_lorentz.txt"

uni=np.loadtxt(file_uniform)

exp=np.loadtxt(file_exp)


lorentz=np.loadtxt(file_lorentz)
```

```python

bins = np.linspace(min(uni[:,0]), max(uni[:,0]), num=100);

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 10))
fig.suptitle(f"Estrazioni da una distribuzione uniforme")

ax1.hist(uni[:,0], bins=bins, histtype="step",density=True,label='$N=1$');
ax1.hist(uni[:,1], bins=bins, histtype="step",density=True,label='$N=2$');
ax1.hist(uni[:,2], bins=bins, histtype="step",density=True,label='$N=10$');
ax1.hist(uni[:,3], bins=bins, histtype="step",density=True,label='$N=100$');

ax1.legend(frameon=True)
ax1.set_xlabel(r"Estrazioni",  fontsize=15)
ax1.set_ylabel(r"Densità di probabilità" ,  fontsize=15)

# best fit of data
(mu, sigma) = norm.fit(uni[:,3])


# add a 'best fit' line
y =  norm.pdf( bins, mu, sigma)
ax2.plot(bins, y, 'b--', linewidth=1, label='Fit gaussiano')
ax2.hist(uni[:,3], bins=bins, histtype="step",density=True,label='$N=100$', color='r');
ax2.set_xlabel(r"Estrazioni",  fontsize=15)
ax2.set_ylabel(r"Densità di probabilità" ,  fontsize=15)

ax2.legend(frameon=True)
ax2.grid(True)



```

```python

bins = np.linspace(min(exp[:,0]), max(exp[:,0]), num=100);

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 10))
fig.suptitle(f"Estrazioni da una distribuzione uniforme")

ax1.hist(exp[:,0], bins=bins, histtype="step",density=True,label='$N=1$');
ax1.hist(exp[:,1], bins=bins, histtype="step",density=True,label='$N=2$');
ax1.hist(exp[:,2], bins=bins, histtype="step",density=True,label='$N=10$');
ax1.hist(exp[:,3], bins=bins, histtype="step",density=True,label='$N=100$');

ax1.legend(frameon=True)
ax1.set_xlabel(r"Estrazioni",  fontsize=15)
ax1.set_ylabel(r"Densità di probabilità" ,  fontsize=15)


# best fit of data
bins = np.linspace(min(exp[:,0]), max(exp[:,0]), num=500);
(mu, sigma) = norm.fit(exp[:,3])
y =  norm.pdf( bins, mu, sigma)
ax2.plot(bins, y, 'b--', linewidth=1, label='Fit gaussiano')
ax2.hist(exp[:,3], bins=bins, histtype="step",density=True,label='$N=100$',color='r');
ax2.set_xlim(0,2)
ax2.legend(frameon=True)
ax2.grid(True)
ax2.set_xlabel(r"Estrazioni",  fontsize=15)
ax2.set_ylabel(r"Densità di probabilità" ,  fontsize=15)
```

```python

bins = np.linspace(-15, 15, num=150);

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 10))
fig.suptitle(f"Estrazioni da una distribuzione Lorentziana")

ax1.hist(lorentz[:,0], bins=bins, histtype="step",density=True,label='$N=1$');
ax1.hist(lorentz[:,1], bins=bins, histtype="step",density=True,label='$N=2$');
ax1.hist(lorentz[:,2], bins=bins, histtype="step",density=True,label='$N=10$');
n, bins, patches  = ax1.hist(lorentz[:,3], bins=bins, histtype="step",density=True,label='$N=100$');

ax1.legend(frameon=True)
ax1.set_xlabel(r"Estrazioni",  fontsize=15)
ax1.set_ylabel(r"Densità di probabilità" ,  fontsize=15)

def cauchy(x,a,b):
    return 1/np.pi*b/((x-a)**2+b**2)


parametri, parametri_err = curve_fit(cauchy,np.linspace(-15, 15, num=149),n)     

y = cauchy(bins,parametri[0],parametri[1])

ax2.plot(bins, y, 'b--', linewidth=1, label='Fit Cauchy')
ax2.hist(lorentz[:,3], bins=bins, histtype="step",density=True,label='$N=100$',color='r');
#ax2.set_xlim(0,1)
ax2.legend(frameon=True)
ax2.grid(True)
ax2.set_xlabel(r"Estrazioni",  fontsize=15)
ax2.set_ylabel(r"Densità di probabilità" ,  fontsize=15)
```
