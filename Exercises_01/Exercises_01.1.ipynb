# Introduzione

In questo esercizio verifichiamo che il generatore di numeri pseudo-casuali funzioni correttamente. In particolare eseguiamo tre controlli distinti: 
* Al posto di considerare tutto il campione contente N=100000 realizzazioni, considerariamo come unità la media di blocchi di dimensione 1000. Abbiamo così un array di 100 elementi. Verifichiamo che il generatore funzioni correttamente, osservando se la media di un numero crescente di elementi dell'array converge al valore atteso 0.5
* Anche per la varianza procediamo in maniera analoga, verificando che la varianza di un numero crescente di elementi dell'array converga al valore atteso 1/12
* Infine eseguiamo un test del $\chi^2$. Per farlo dividiamo il nostro intervallo $[0,1]$ in 100 sottointervalli e, generando 10000 numeri pseudo-casuali, vediamo come questi si distribuiscono nei 100 bin. In particolare, ci aspettiamo di ottenere un chi-quadro di 100. Ripetiamo poi questa procedura N=100 volte.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import string
from scipy.optimize import curve_fit

import seaborn as sns
sns.set_theme(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1.2, color_codes=True, rc=None)
```

```python
# Carico i dati dai file

file_mean= "Exercises_01.1/file_mean.txt"
file_sigma= "Exercises_01.1/file_sigma.txt"
file_chi_quadro = "Exercises_01.1/file_chi_quadro.txt"

mean_data, std_mean =np.loadtxt(file_mean,delimiter = ",", unpack = True)
number_throws = np.shape(mean_data)[0]
sigma_data, std_sigma =np.loadtxt(file_sigma,delimiter = ",", unpack = True)
chi_quadro_data = np.loadtxt(file_chi_quadro)
```

```python
#Rappresento graficamente i risultati ottenuti

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 20))

ax1.errorbar(
    np.arange(1,(number_throws+1))*1000,
    mean_data-0.5,
    yerr = std_mean,  
    marker='o', linestyle=""
)

ax1.grid(True)

ax1.set_title(f"Estrazioni da una distribuzione uniforme",  fontsize=20);

ax1.set_xlabel(r"Numero di estrazioni",  fontsize=15)
ax1.set_ylabel(r"$ \langle r \rangle -\frac{1}{2}  $" ,  fontsize=15)



ax2.errorbar(
    np.arange(1,number_throws+1)*1000,
    sigma_data-1/12,
    yerr = std_sigma,
    marker='o', linestyle=""
    
)
ax2.grid(True)


ax2.set_xlabel(r"Numero di estrazioni",  fontsize=15)
ax2.set_ylabel(r"$ \left\langle \left(r \rangle -\frac{1}{2}\right)^2 \right\rangle-\frac{1}{12}  $" ,  fontsize=15)

ax3.plot(
    np.arange(0,100),
    chi_quadro_data,
    label='Valore misurato',
    marker='o', linestyle=""
)
ax3.grid(True)


ax3.set_xlabel(r"Numero di ripetizioni della misura",  fontsize=15)
ax3.set_ylabel(r"$ \chi^2  $" ,  fontsize=15)
ax3.hlines(100,0,100, colors='red', linestyles='dashed', label='Valore atteso')

ax3.legend(frameon=True)
```


>Sebbene la $\langle r \rangle $ e la $\left\langle \left(r \rangle -\frac{1}{2}\right)^2 \right\rangle$ convergano al valore atteso, osserviamo che il test del $\chi^2$ mostra grandi fluttuazioni. Questo ci suggerisce che il nostro generatore di numeri pseudo-casuali non sia troppo attendibile e che esso possa essere migliorato.
