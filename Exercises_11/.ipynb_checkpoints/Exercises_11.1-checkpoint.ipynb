# Introduzione

In questa esercitazione implementiamo un algortimo di supervised machine learning per fittare un set di dati. La procedura standard che seguiamo è la seguente: 

1. Prepariamo due set indipendenti di dati da usare come data-training e data-validation. Il primo serve per "allenare" l'algoritmo mentre il secondo viene utilizzato per valutare le performance dell'algortimo. 

        x_train = np.random.uniform(-1, 1, 100)
        x_valid = np.random.uniform(-1, 1, 50)

        y_target = m * x_valid + b 
        y_train = np.random.normal(m * x_train + b, sigma) 
        y_valid = np.random.normal(m * x_valid + b, sigma)


2. Costruiamo il network di nodi e legami fra i diversi layer dell'algoritmo:

        model = tf.keras.Sequential()
        model.add(Dense(1, input_shape=(1,)))
        

3. Scegliamo i parametri dinamici dei modello, ossia quei parametri che specificano come esso si muove nello spazio dei valori possibili. Questo richiede di scegliere tre diverse funzioni:

    * una **funzione di attivazione** che descrive l'output di ogni neurone al variare dell'input.
    * una **funzione costo** che l'algoritmo minimizza nella ricerca dei parametri migliori per il modello usando solo il data-training. 
    * un **algoritmo di minimizzazione** che descriva il modo in cui il sistema si muove nello spazio dei parametri.
    * una **metrica** che determini il grado di precisione del nostro algoritmo ottenuto confrontando i risultati del data-validation con i risultati reali. 

            model.compile(activation='sigmoid', optimizer='sgd', loss='mse', metrics=['mse']);
            
   
3. Eseguiamo il training dell'algoritmo sui data-test. Il modo ottimale in cui questo si può fare è aggiornare il network ogni `batch_size` dati processati piuttosto che dopo ogni dato ed rieseguire la procedura completa `epochs` volte su tutto il data set. Infine misuriamo l'accuratezza del risultato ottenuto attraverso una procedura di validazione sui `data-valid`.

        model.fit(x=x_train, y=y_train, 
                  batch_size=32, epochs=10,
                  validation_data=(x_valid, y_valid)
                  );



# Linear fit

Per prima cosa fittiamo una funzione lineare variando i parametri del modello. In questo caso specifico non serve utilizzare una particolare funzione di attivazione poiché un network base ha già una struttura lineare (network value, weight, bias). Usiamo inoltre un solo layer composto da un solo neurone dal momento che vogliamo descrive un modello lineare. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import string
from scipy.optimize import curve_fit
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stat
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import subprocess as sb
from os import system
import os
import shutil
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

sns.set_theme(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1.2, color_codes=True, rc=None)
```

Per prima cosa variamo il numero di epoche $N_{epochs}$ ed il numero di punti per ogni epoca $N_{train}$.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

# generate training inputs
N_epochs =[10,20,30]
N_train = [50,500,5000]

l=0
for i in N_epochs:
    fig, ax = plt.subplots(3, 2, figsize=(15,15));  
    fig.patch.set_facecolor(color=plt.cm.tab10(l))
    fig.patch.set_alpha(0.1)
    plt.suptitle(r'$N_{epochs}=$'+str(i))
    ax= ax.ravel()
    l=l+1
    k=0
    for j in N_train:
        np.random.seed(0)
        x_train = np.random.uniform(-1, 1, j)
        x_valid = np.random.uniform(-1, 1, 50)
        x_valid.sort()
        y_target = m * x_valid + b # ideal (target) linear function

        sigma = 0.0 # noise standard deviation, for the moment it is absent
        y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
        y_valid = np.random.normal(m * x_valid + b, sigma)


        model = tf.keras.Sequential()
        model.add(Dense(1, input_shape=(1,)))

        # compile the model choosing optimizer, loss and metrics objects
        model.compile(optimizer='sgd', loss='mse', metrics=['mse']);

        history = model.fit(x=x_train, y=y_train, 
                  batch_size=32, epochs=i,
                  shuffle=True, # a good idea is to shuffle input before at each epoch
                  validation_data=(x_valid, y_valid),
                  verbose=0);


        # look into training history
        # Plot training & validation loss values
        ax[k].plot(range(1, i + 1),history.history['loss'])
        ax[k].plot(range(1, i + 1),history.history['val_loss'])
        ax[k].set_title(r'Model loss for $N_{train} =$'+str(j))
        ax[k].set_ylabel('Loss')
        ax[k].set_xlabel('Epoch')
        ax[k].legend(['Train', 'Test'], loc='best')
        
        k= k+1

        x_predicted = np.random.uniform(-1, 1, 100)
        y_predicted = model.predict(x_predicted)
        ax[k].scatter(x_predicted, y_predicted,color='r',label='Predected')
        ax[k].plot(x_valid, y_target,label='Theoretical')
        ax[k].set_title('Fit for $N_{train} =$'+str(j))
        ax[k].grid(True)
        ax[k].set_xlabel('x')
        ax[k].set_ylabel('y')
        ax[k].legend(loc='best')
        
        k= k+1
        
        plt.tight_layout()
        plt.subplots_adjust(wspace=0.3)
```

Vediamo come 10 epoche non siano sufficienti per ottenere risulati soddisfacenti. Per $N_{epochs}=20$ e $N_{epochs}=30$ abbiamo invece dei buoni risultati soprattutto quando $N_{train}=500$ o $N_{train}=5000$. In particolare ilcompromesso ottimale sembra essere 

$$ \boxed{N_{epochs = 30} \quad \quad N_{train}=500}$$

Chiaramente nel caso precedente non si presentava alcun problema di overfitting per il fatto che i dati non avevano rumore. Valutiamo ora l'effetto della rumorosità dei dati sulla convergenza del fit.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

# generate training inputs
sigma = [0.5,0.75,1,1.25]

l=0
for i in sigma:
    k=0
    fig, ax = plt.subplots(1, 2, figsize=(18,6)); 
    fig.patch.set_facecolor(color=plt.cm.tab10(l))
    fig.patch.set_alpha(0.1)
    plt.suptitle(r'$sigma=$'+str(i))
    ax= ax.ravel()
    l=l+1
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, 500)
    x_valid = np.random.uniform(-1, 1, 50)
    x_valid.sort()
    y_target = m * x_valid + b # ideal (target) linear function

    sigma = i 
    y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma)


    model = tf.keras.Sequential()
    model.add(Dense(1, input_shape=(1,)))

    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse']);

    history = model.fit(x=x_train, y=y_train, 
              batch_size=32, epochs=30,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid),
              verbose=0);


    # look into training history
    # Plot training & validation loss values
    ax[k].plot(history.history['loss'])
    ax[k].plot(history.history['val_loss'])
    ax[k].set_title(r'Model loss')
    ax[k].set_ylabel('Loss')
    ax[k].set_xlabel('Epoch')
    ax[k].legend(['Train', 'Test'], loc='best')

    k= k+1

    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    ax[k].scatter(x_predicted, y_predicted,color='r',label='Predected')
    ax[k].plot(x_valid, y_target,label='Theoretical')
    ax[k].scatter(x_valid, y_valid, marker='.', label='Validation data')
    ax[k].set_title('Fit')
    ax[k].grid(True)
    ax[k].set_xlabel('x')
    ax[k].set_ylabel('y')
    ax[k].legend(loc='best')

    
```

> Attreverso questa analisi vediamo chiaramente che la rumorosità dei dati non influenza particolarmente la convergenza dell'algoritmo. Inoltre vediamo che non si presenta alcun problema di overfitting dal momento che la loss per i data di test è sempre maggiore rispetto a quella dei dati di training.
