# Introduzione

In questo caso vogliamo fittare una funzione che non è più lineare:
$$
f(x)=4-3x-2x^2+3x^3
$$
con $x \in [-1,1]$.
Per fare questo dobbiamo quindi introdurre più di un layer nella nostra rete neaurale ed una funzione di attivazione non-lineare che permetta di descrivere il risultato della funzione $f(x)$. Poichè a priori non è possibile identificare quale sia la configurazione ideale della rete, avanziamo diverse proposte e vediamo che tipo di risultato otteniamo. In particolare scegliamo di lavorare con questo ordine:

1. Per prima cosa selezioniamo quale sia la migliore funzione di attivazione.

2. Dopodiché selezioniamo il compromesso più favorevole fra $N_{train}$ e $N_{epochs}$.

3. Consideriamo poi come incide il numero di neuroni e layers nell'ottenere il fit ottimale.

4. Infine valutiamo altre funzioni di ottimizzazione e di loss.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import string
from scipy.optimize import curve_fit
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stat
import pandas as pd
import subprocess as sb
from os import system
import os
import shutil
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

sns.set_theme(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1.2, color_codes=True, rc=None)
```

```python
def func(x):
    return 4 -3 * x -2*x**2 + 3*x**3

x_valid = np.random.uniform(-1, 1, 500)
x_valid.sort()
y_target = func(x_valid) 
```

```python
def machine_learning(activation, n_epochs, n_train, n_layers, n_neurons, optimizer, loss):

    a=0
    for k in activation:
        for i in n_layers:
            for j in n_neurons:
                for m in n_epochs:
                    for n in n_train:
                        for p in optimizer:
                            for l in loss:
                                np.random.seed(0)
                                x_train = np.random.uniform(-1, 1, n)
                                x_valid = np.random.uniform(-1, 1, m)
                                x_valid.sort()
                                y_target = func(x_valid) # ideal (target) linear function
                                sigma = 0.2 # noise standard deviation, for the moment it is absent
                                y_train = np.random.normal(func(x_train), sigma) # actual measures from which we want to guess regression parameters
                                y_valid = np.random.normal(func(x_valid), sigma)
        
                                model = keras.models.Sequential()

                                # 1 layer
                                model.add(keras.layers.Dense(
                                    j,
                                    activation=k,
                                    input_shape=(1,)))

                                # layer nascosti
                                for r in range(1, i):
                                    model.add(keras.layers.Dense(
                                            j,
                                            activation=k))

                                # layer finale
                                model.add(keras.layers.Dense(
                                    1,
                                    activation=k))  

                                model.compile(
                                    loss=l,
                                    optimizer=p,
                                    metrics=['mse']
                                    )


                                history = model.fit(x=x_train, y=y_train, 
                                                    batch_size=32, epochs=m,
                                                    verbose=0,
                                                    validation_data=(x_valid, y_valid),
                                                    shuffle=True)
                                
                                x_predicted = np.random.uniform(-1, 1, m)

                                df_1 = pd.DataFrame({'activation':k,'n_layers': i, 'n_neurons': j,'n_epochs': m, 'n_train':n, 'optimizer': p,
                                                       'loss': l, 'history_loss': np.transpose(history.history['loss']), 'history_val_loss': np.transpose(history.history['val_loss']),
                                                       'x_predicted': x_predicted, 'y_predicted': model.predict(x_predicted).flatten('F')})

                                if(a==0):
                                    df = df_1
                                else:
                                    df= df.append(df_1)
                                
                                a=a+1
                                
                
    return df
```

## 1. Funzione di attivazione

Per identificare quale sia la migliore funzione di attivazione, proviamo sia funzioni open-bounded che funzioni con plateu.

```python
activation = ['tanh','sigmoid','softsign', 'elu', 'relu','selu','gelu']
n_epochs = [100]
n_train = [1000]
n_neuron = [10]
n_layer = [5]
optimizer = ['sgd']
loss = ['mse']

df = machine_learning(activation, n_epochs, n_train , n_layer, n_neuron, optimizer, loss)
```

```python
m=0
for i in activation:      
    fig, ax = plt.subplots(1, 2, figsize=(18,6));  
    fig.patch.set_facecolor(color=plt.cm.Pastel2(m))
    fig.patch.set_alpha(0.3)
    plt.suptitle('Activation function: '+str(i))
    fig.subplots_adjust(top=0.9)
    ax= ax.ravel()
    k=0
    m=m+1

    mask = (df.activation==i) 

    ax[k].plot(df[mask].history_loss.to_numpy())
    ax[k].plot(df[mask].history_val_loss.to_numpy())
    ax[k].set_ylabel('Loss')
    ax[k].set_xlabel('Epoch')
    ax[k].set_yscale('log')
    ax[k].legend(['Train', 'Test'], loc='best')

    k=k+1

    ax[k].scatter(df[mask].x_predicted.to_numpy(), df[mask].y_predicted.to_numpy(),color='r',label='Predected')
    ax[k].plot(x_valid, y_target,label='Theoretical')
    ax[k].grid(True)
    ax[k].set_xlabel('x')
    ax[k].set_ylabel('y')
    ax[k].legend(loc='best')
    k= k+1
```

Le funzioni di attivazione più adatte sembrano essere la funzione **relu** e la funzione **gelu**. Infatti la prima raggiunge il livello di loss inferiore ma è particolarmente irregolare sia per l'insieme di train sia per quello di test. La seconda invece raggiunge buoni livelli di loss ma ha un comportamento più stabile che suggerisce un'ulteriore possibile miglioramento del fit. Osserviamo infine che tutte le funzioni di attivazione upper-bounded non sono adatta a fittare questa funzione.

## 2. Numero di epoche e di train



Vogliamo ora identificare quale sia il numero ottimale di valori di trainig e di epoche da fornire all'algoritmo. Chiaramente il risultato che otteniamo dipende fortemente dalla scelta del numero di neuroni e di layer. Infatti aumentando $N_{neurons}$ e $N_{layer}$ aumentiamo il numero di parametri da ottimizzare. Tuttavia siamo interessati ad una stima approssimativa per un numero intermedio di $N_{neurons}$ e $N_{layer}$ e quindi scegliamo le stesse condizioni del caso precedente.

```python
activation = ['relu','gelu']
n_epochs = [10,50,100,200]
n_train = [500, 1000, 1500, 2000]
n_neuron = [10]
n_layer = [5]
optimizer = ['sgd']
loss = ['mse']

df = machine_learning(activation, n_epochs, n_train , n_layer, n_neuron, optimizer, loss)
```

```python
m=0
for i in n_epochs:
    for j in n_train:        
        fig, ax = plt.subplots(1, 4, figsize=(22,5));  
        fig.patch.set_facecolor(color=plt.cm.tab20c(m))
        fig.patch.set_alpha(0.3)
        plt.suptitle(r'$N_{epochs}=$ '+str(i)+r'$ \quad N_{train} =$ '+str(j))
        fig.subplots_adjust(top=0.8)
        ax= ax.ravel()
        k=0
        m=m+1
        for q in activation:
            mask = (df.n_epochs==i) & (df.n_train==j) & (df.activation==q)

            ax[k].plot(df[mask].history_loss.to_numpy())
            ax[k].plot(df[mask].history_val_loss.to_numpy())
            ax[k].set_ylabel('Loss')
            ax[k].set_xlabel('Epoch')
            ax[k].set_yscale('log')
            ax[k].legend(['Train', 'Test'], loc='best')
            ax[k].set_title('Activation function: '+str(q))

            k=k+1

            ax[k].scatter(df[mask].x_predicted.to_numpy(), df[mask].y_predicted.to_numpy(),color='r',label='Predected')
            ax[k].plot(x_valid, y_target,label='Theoretical')
            ax[k].grid(True)
            ax[k].set_xlabel('x')
            ax[k].set_ylabel('y')
            ax[k].legend(loc='best')
            ax[k].set_title('Activation function: '+str(q))
            k= k+1
```

Da questa analisi osserviamo che:

* La funzione di attivazione gelu è molto più stabile rispetto alla relu e questo le consente di convergere anche per valori minori di $N_{epochs}$ e $N_{train}$. Scegliamo quindi questa funzione per proseguire la nostra analisi.

* La funzione relu ha uno strano comportamento per alcuni valori di $N_{epochs}$ e $N_{train}$. Non siamo riusciti a capire dove possa essere il problema.

Concludiamo quindi che: 

$$ \boxed{\text{Funzione di attivazione: gelu} \quad \quad N_{epochs} = 200 \quad \quad N_{train} = 1000}$$


## 3. Numero di layer e neuroni

Ora cerchiamo quale sia il numero ottimale di layers e di neuroni avendo già fissato $N_{epochs}$ e $N_{train}$. In particolare scegliamo di lavorare con $2\leq N_{layers}\leq 8$ e $5 \leq N_{neurons} \leq 20$.

```python
activation = ['gelu']
n_epochs = [200]
n_train = [1000]
n_neuron = [5,10,15,20]
n_layer = [2,4,6,8]
optimizer = ['sgd']
loss = ['mse']

df = machine_learning(activation, n_epochs, n_train , n_layer, n_neuron, optimizer, loss)
```

```python
m=0
for i in n_layer:
    for j in n_neuron:        
        fig, ax = plt.subplots(1, 2, figsize=(18,6));  
        fig.patch.set_facecolor(color=plt.cm.tab20c(m))
        fig.patch.set_alpha(0.3)
        plt.suptitle(r'$N_{layer}=$ '+str(i)+r'$ \quad N_{neurons} =$ '+str(j))
        fig.subplots_adjust(top=0.9)
        ax= ax.ravel()
        k=0
        m=m+1
        
        mask = (df.n_layers==i) & (df.n_neurons==j)
        
        ax[k].plot(df[mask].history_loss.to_numpy())
        ax[k].plot(df[mask].history_val_loss.to_numpy())
        ax[k].set_ylabel('Loss')
        ax[k].set_xlabel('Epoch')
        ax[k].set_yscale('log')
        ax[k].legend(['Train', 'Test'], loc='best')

        k=k+1

        ax[k].scatter(df[mask].x_predicted.to_numpy(), df[mask].y_predicted.to_numpy(),color='r',label='Predected')
        ax[k].plot(x_valid, y_target,label='Theoretical')
        ax[k].grid(True)
        ax[k].set_xlabel('x')
        ax[k].set_ylabel('y')
        ax[k].legend(loc='best')
        k= k+1
```

Osserviamo come aumentare $N_{layers}$ e $N_{neurons}$ consenta di ottenere risultati sempre migliori. In particolare per $N_{layers}=8$ e $N_{neurons}=20$ vediamo che anche l'estremo del dominio ($x\approx 1$) viene fittato correttamente dal nostro algortimo. Tuttavia, a parte questo, non si osserva un significativo aumento dell'efficienza per valori intermedi di $N_{layers}$ e $N_{neurons}$. Pertanto la configurazione ottimale della rete è: 

$$ \boxed{\text{Funzione di attivazione: gelu} \quad \quad N_{epochs} = 200 \quad \quad N_{train} = 1000 \quad \quad N_{layers}=6 \quad \quad N_{neurons}=5}$$

## 4. Funzioni di ottimizzazione e di loss

```python
activation = ['selu']
n_epochs = [200]
n_train = [1000]
n_neuron = [5]
n_layer = [6]
optimizer = ['sgd','adam']
loss = ['mse','binary_crossentropy']

df = machine_learning(activation, n_epochs, n_train , n_layer, n_neuron, optimizer, loss)
```

Vediamo chiaramente che tra le due funzione di loss utilizzate, l'unica ragionevole è l'mse mentre per quanto la funzione di ottimizzazione non è riscontrabile una particolare diversità. In conclusione la rete neurale ottimale per fittare questa funzione è: 


$$ \boxed{\text{Funzione di attivazione: selu} \quad \quad \text{Funzione di loss: mse} \quad \quad \text{Funzione di ottimizzazione: sgd}} $$
$$\boxed{ N_{epochs} = 200 \quad \quad N_{train} = 1000 \quad \quad N_{layers}=6 \quad \quad N_{neurons}=5  }$$

```python
def machine_learning(activation, n_epochs, n_train, n_layers, n_neurons, optimizer, loss):

    a=0
    for k in activation:
        for i in n_layers:
            for j in n_neurons:
                for m in n_epochs:
                    for n in n_train:
                        for p in optimizer:
                            for l in loss:
                                np.random.seed(0)
                                x_train = np.random.uniform(-1, 1, n)
                                x_valid = np.random.uniform(-3, 3, m)
                                x_valid.sort()
                                y_target = func(x_valid) # ideal (target) linear function
                                sigma = 0.2 # noise standard deviation, for the moment it is absent
                                y_train = np.random.normal(func(x_train), sigma) # actual measures from which we want to guess regression parameters
                                y_valid = np.random.normal(func(x_valid), sigma)
        
                                model = keras.models.Sequential()

                                # 1 layer
                                model.add(keras.layers.Dense(
                                    j,
                                    activation=k,
                                    input_shape=(1,)))

                                # layer nascosti
                                for r in range(1, i):
                                    model.add(keras.layers.Dense(
                                            j,
                                            activation=k))

                                # layer finale
                                model.add(keras.layers.Dense(
                                    1,
                                    activation=k))  

                                model.compile(
                                    loss=l,
                                    optimizer=p,
                                    metrics=['mse']
                                    )


                                history = model.fit(x=x_train, y=y_train, 
                                                    batch_size=32, epochs=m,
                                                    verbose=0,
                                                    validation_data=(x_valid, y_valid),
                                                    shuffle=True)
                                
                                x_predicted = np.random.uniform(-3, 3, m)

                                df_1 = pd.DataFrame({'activation':k,'n_layers': i, 'n_neurons': j,'n_epochs': m, 'n_train':n, 'optimizer': p,
                                                       'loss': l, 'history_loss': np.transpose(history.history['loss']), 'history_val_loss': np.transpose(history.history['val_loss']),
                                                       'x_predicted': x_predicted, 'y_predicted': model.predict(x_predicted).flatten('F'), 'x_valid':x_valid,'y_target':y_target })

                                if(a==0):
                                    df = df_1
                                else:
                                    df= df.append(df_1)
                                
                                a=a+1
                                
                
    return df
```
